# line plot with seaborn


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tips.csv")
```

```python
sns.lineplot(df['total_bill'],marker='^')
```

# BAR PLOT

```python
sns.barplot(x=df['day'],y=df['total_bill'])
```

# FOR CHECKING THE DISTRIBUTION

```python
sns.displot(df['total_bill'],kind='kde',fill=True)
```

# IF ANY DISPLOT GIVE BELL CURVE THAN THE DISTRIBUTION IS EVENLY SPREAD(NORMAL DISTRIBUTION)

```python
sns.barplot(df['total_bill'])
```

```python
dinner=df[df['time']=='Dinner']
lunch=df[df['time']=='Lunch']
```

```python
sns.barplot(dinner['total_bill'])
```

```python
sns.barplot(lunch['total_bill'])
```

```python
sns.histplot(dinner['total_bill'])
```

```python
label=list(df['day'].value_counts().index)
value=list(df['day'].value_counts().values)
```

```python
plt.pie(value,labels=label,explode=[0,0.2,0,0.2],colors=['r','y','g','b'],autopct="%1.2f%%",shadow=True,wedgeprops={"ec":"k"})
plt.show()
```
