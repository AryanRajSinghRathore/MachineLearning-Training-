```python
correlation=df2.corr()
correlation
```

```python
correlation['price']
```

```python
correlation[correlation['price'] > 0.10]
```

```python
column_list=list(correlation[correlation['price'] > 0.10].index)
column_list
```

```python
df3=df2[column_list]
df3.head()
```

```python
df3.isnull().sum()
```

```python
x=df3.drop('price',axis=1)
x
```

```python
y=df3[['price']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
dtr.score(x_train,y_train)
```

```python
dtr.score(x_test,y_test)#decision tree leads to overfit 
```

```python
pred=dtr.predict(x_test)
```

```python
compare=y_test
```

```python
compare['prediction']=pred
```

```python
compare['prediction']=compare['prediction'].apply(lambda x:'{:.2f}'.format(x))
```

```python
compare
```

```python
y_test=y_test.drop('prediction',axis=1)
y_test
```
